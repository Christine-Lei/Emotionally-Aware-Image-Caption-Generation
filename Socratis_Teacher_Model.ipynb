```python
ethnicity_cols = ['White', 'Black/African-American', 'Asian/Pacific Islander', 'Hispanic/Latino', 'Native American', 'Other']
```

```python
def flatten_entry(data_row):
  all_data = []
  for data in data_row:
    flattened_data = []

    for i, item in enumerate(data):
        if isinstance(item, list):
            flattened_data.extend(item)
        elif isinstance(item, dict):
            flattened_data.extend(item.values())
        else:
            flattened_data.append(item)
    # one hot encode ethnicities
    ethnicity_dict = {i:0 for i in ethnicity_cols}
    if flattened_data[-1]:
      for i in flattened_data[-1][0]:
        # for true part
        if i in ethnicity_cols:
          ethnicity_dict[i] = 1
    full_dict = {col:flattened_data[i] for i, col in enumerate(columns)}
    full_dict.update(ethnicity_dict)
    del full_dict['ethnicity']
    all_data.append(full_dict)
  return pd.DataFrame(all_data)
```

```python
def process_labels(label_list):
    processed_labels = []
    for idx, label in enumerate(label_list):
        new_label = label
        if len(label) == 1:
            continue
        if " " in label:
            # if more than one word remove it
            continue
        if '"' in label:
            new_label = label.strip('"')
        if "(" in label:
            par_idx = label.index("(")
            new_label = label[:par_idx]
        if new_label not in processed_labels:
            processed_labels.append(new_label)
    return processed_labels
```

```python
import os
import numpy as np
import pandas as pd

# torch
import torch
from torch.utils.data import Dataset
from transformers import BertTokenizer

# from visualbert
from processing_image import Preprocess
from utils import Config
from modeling_frcnn import GeneralizedRCNN
```

inspired by [this link](https://gist.github.com/ji1kang/6cfa1810d754c3087ef3ed4b95ee6c27)

```python
class ImageProcessor:
    def __init__(self, device='cuda'):
        frcnn_cfg = Config.from_pretrained("unc-nlp/frcnn-vg-finetuned")
        frcnn_cfg.MODEL.DEVICE = device
        self.device = device

        self.frcnn = GeneralizedRCNN.from_pretrained("unc-nlp/frcnn-vg-finetuned", config=frcnn_cfg)

        self.frcnn_cfg = frcnn_cfg
        self.image_preprocess = Preprocess(frcnn_cfg)

    def get_visual_embeddings(self, image_path):
        # run frcnn
        images, sizes, scales_yx = self.image_preprocess(image_path)

        output_dict = self.frcnn(
            images,
            sizes,
            scales_yx=scales_yx,
            padding="max_detections",
            max_detections=self.frcnn_cfg.max_detections,
            return_tensors="pt",
        )
        features = output_dict.get("roi_features").detach().cpu()
        return features
```

```python
encoded_labels_list = []
for idx in range(len(df)):
    emotions = df['emotions'][idx]
    encoded_labels = np.zeros(len(label_map), dtype=np.float32)
    for emotion in emotions:
        encoded_labels[label_map[emotion]] = 1
    encoded_labels_list.append(encoded_labels)

encoded_labels_tensor = torch.tensor(np.array(encoded_labels_list))
```

```python
class SocratisDataset(Dataset):
    def __init__(self, data, images_base_path, encoded_labels, device='cuda'):
        self.data = data
        self.images_base_path = images_base_path

        # feature extractors
        self.tokenizer = BertTokenizer.from_pretrained("google-bert/bert-base-uncased")
        self.visual_extractor = ImageProcessor(device=device)
        self.encoded_labels = encoded_labels

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        captions, image_name = self.data.iloc[idx]['caption'], self.data.iloc[idx]['image_name']

        # get image embedings
        image_path = os.path.join(self.images_base_path, image_name)
        visual_embeds = self.visual_extractor.get_visual_embeddings(image_path)
        visual_token_type_ids = torch.ones(
            visual_embeds.shape[:-1], dtype=torch.long)
        visual_attention_mask = torch.ones(
            visual_embeds.shape[:-1], dtype=torch.float)

        # get text embeddings
        inputs = self.tokenizer(captions, return_tensors="pt", max_length=32, truncation=True, padding='max_length')
        input_ids = inputs["input_ids"]
        token_type_ids = inputs["token_type_ids"]
        attention_mask = inputs["attention_mask"]


        labels = self.encoded_labels[idx]

        return (input_ids, token_type_ids, attention_mask,
                visual_embeds, visual_token_type_ids, visual_attention_mask,
                labels)
```

```python
# Initialize dataset and dataloader
images_path = './images'
socratis_dataset = SocratisDataset(df, images_path, encoded_labels_tensor)
dataloader = DataLoader(socratis_dataset, batch_size=64, shuffle=True)
```

```python
# Define teacher model architecture
class TeacherModel(nn.Module):
    def __init__(self, visualbert_model):
        super(TeacherModel, self).__init__()
        self.visualbert = visualbert_model
        self.fc = nn.Linear(self.visualbert.config.hidden_size, num_emotions)

    def forward(self, input_ids, token_type_ids, attention_mask, visual_embeds, visual_token_type_ids, visual_attention_mask, labels):
        visualbert_outputs = self.visualbert(input_ids=input_ids.squeeze(1),
                                             attention_mask=attention_mask.squeeze(1),
                                             token_type_ids=token_type_ids.squeeze(1),
                                             visual_embeds=visual_embeds.squeeze(1),
                                            visual_token_type_ids=visual_token_type_ids.squeeze(1),
                                            visual_attention_mask=visual_attention_mask.squeeze(1))
        pooled_output = visualbert_outputs['pooler_output']

        # Emotion prediction
        logits = self.fc(pooled_output) # Loss function operates from logits

        # probabilities = F.softmax(logits, dim=-1)
        return logits
```

```python
visualbert_model = VisualBertModel.from_pretrained("uclanlp/visualbert-vqa-coco-pre") # freeze --> training last layers
```

```python
for param in visualbert_model.parameters():
    param.requires_grad = False
```

```python
visualbert_model.config
```

```python
# Initialize and instantiate the teacher model
teacher_model = TeacherModel(visualbert_model)
```

```python
num_epochs = 50

optimizer = optim.Adam(teacher_model.parameters(), lr=1e-2)
criterion = nn.BCEWithLogitsLoss()
```

```python
from tqdm import tqdm
```

```python
checkpoint_path = "teacher_checkpoint.pth.tar"
```

```python
import pickle

log_file_path = './log.pickle'
```

```python
def save_checkpoint(model, optimizer, epoch, loss):
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'loss': loss,
    }, checkpoint_path)

    with open(log_file_path, 'wb') as f:
        pickle.dump(log, f)
```

```python
def load_checkpoint(model, optimizer):
    if os.path.exists(checkpoint_path):
        checkpoint = torch.load(checkpoint_path)
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        start_epoch = checkpoint['epoch'] + 1
        loss = checkpoint['loss']
        print(f"Checkpoint loaded. Resuming from epoch {start_epoch}")

        with open(log_file_path, 'rb') as f:
            log = pickle.load(f)
        return start_epoch, loss, log
    else:
        print("No checkpoint found. Starting from epoch 1")
        return 0, None, {'epoch':[], 'loss':[]}
```

```python
start_epoch, previous_loss, log = load_checkpoint(teacher_model, optimizer)
```

```python
start_epoch, previous_loss, log
```

```python
# Train teacher model
teacher_model.train()

for epoch in range(start_epoch, num_epochs):
    if previous_loss == 0:
        running_loss = 0.0
    else:
        running_loss = previous_loss

    # Wrap your dataloader with tqdm
    for batch in tqdm(dataloader, desc=f"Epoch {epoch}", leave=False, dynamic_ncols=True):
        input_ids, token_type_ids, attention_mask, visual_embeds, visual_token_type_ids, visual_attention_mask, labels = batch

        # Zero the parameter gradients
        optimizer.zero_grad()

        # Forward pass through teacher model
        probabilities = teacher_model(input_ids, token_type_ids, attention_mask, visual_embeds, visual_token_type_ids, visual_attention_mask, labels)

        # Compute loss
        loss = criterion(probabilities, labels)

        # Backward pass and optimize
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    # Print epoch loss
    print(f"Epoch {epoch}, Loss: {running_loss / len(dataloader)}")

    save_checkpoint(teacher_model, optimizer, epoch, running_loss)

    # log loss and determine when to stop
    log['epoch'].append(epoch+1)
    log['loss'].append(running_loss / len(dataloader))
```
